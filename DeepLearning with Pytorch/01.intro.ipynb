```python
%%time
a = torch.rand(103,103)

value = 0

for i in range(len(a)):
    for j in range(len(a)):
        value += a[i][j]

print(value)
```

```python
%%time
value =0
a = torch.rand(103,103)
value += torch.matmul(a, a)
print(value)
```
